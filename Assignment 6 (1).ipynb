```python

import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
# importing data using .read_csv() function
df = pd.read_csv('Data-Week11.csv')
# printing DataFrame
df
```

```python
# Creating a instance of label Encoder.
le = LabelEncoder()

# Using .fit_transform function to fit label
# encoder and return encoded label
df['Destination_encoded'] = le.fit_transform(df['Destination'])
output_file_path = 'modified_data.csv'
df.to_csv(output_file_path, index=False)
print("Data with numeric 'Destination' saved to:", output_file_path)
```

```python
df1 = pd.read_csv('modified_data.csv', delimiter=",")
X1 = df1[['Destination_encoded']].values
X2 = df1[['price']].values
X=X1+X2
df1.head()
```

```python
labels = range(1, 20)
plt.figure(figsize=(10, 7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X1,X2, label='True Position')

for label, x, y in zip(labels, X1, X2):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

Number of destinations k = 6, from the above graph  0, 1, 2, 3, 4, 5 which represents the 6 cities.

0:'Mumbai', 1: 'Hyderabad', 2: 'Chennai', 3: 'Bangalore', 4: 'Kolkata', 5: 'Delhi'

-> The vertical distribution of points for each destination indicates the range of prices for flights going to that destination. Tighter clusters suggest less variability in price, whereas wider clusters indicate greater variability.
